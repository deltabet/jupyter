```python
import numpy as np
import pandas as pd

train_genre_df = pd.read_csv('trainGenre.csv')
train_genre_df.head()
```

```python
train_genre_df.shape
```

```python
new_train_genre_df = train_genre_df.drop('UserID', axis=1)
new_train_genre_df = new_train_genre_df.drop('MovieID', axis=1)
new_train_genre_df.head()
```

```python
# figure out where the NaN values are and drop their rows
import math
print new_train_genre_df.shape

drop_indices = []
for ind, row in new_train_genre_df.iterrows():
    if math.isnan(row['Rating']):
        drop_indices.append(ind)
print drop_indices
```

```python
#check is theres any NaN values
new_train_genre_df = new_train_genre_df.drop(new_train_genre_df.index[drop_indices])
print new_train_genre_df.isnull().values.any()
```

```python
cols = new_train_genre_df.columns
cols = cols.drop('Unnamed: 0')
feature_cols = cols.drop('Rating')
print feature_cols

X = new_train_genre_df[feature_cols]
Y = new_train_genre_df['Rating']

from sklearn.cross_validation import train_test_split
from sklearn import metrics
from sklearn.linear_model import LinearRegression
mlreg = LinearRegression()
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=1)
mlreg.fit(X_train, Y_train)
Y_pred = mlreg.predict(X_test)
```

```python
print np.sqrt(metrics.mean_squared_error(Y_test, Y_pred))
print metrics.mean_squared_error(Y_test, Y_pred)
```

```python
from sklearn import preprocessing
users = pd.read_csv("users_avg.csv")
del users['Zip-code']
del users['Average']
le = preprocessing.LabelEncoder()
le.fit(['F', 'M'])
le2 = preprocessing.LabelEncoder()
le2.fit([1, 18, 25, 35, 45, 50, 56])
gendersNew = le.transform(users['Gender'].as_matrix())
users['Gender'] = gendersNew
ageNew = le2.transform(users['Age'].as_matrix())
users['Age'] = ageNew
users.head()
```

```python
dropColumns = ['G_F', 'G_M', 'A_1', 'A_18', 'A_25', 'A_35', 'A_45', 'A_50',
       'A_56', 'O_0', 'O_1', 'O_2', 'O_3', 'O_4', 'O_5', 'O_6', 'O_7',
       'O_8', 'O_9', 'O_10', 'O_11', 'O_12', 'O_13', 'O_14', 'O_15',
       'O_16', 'O_17', 'O_18', 'O_19', 'O_20']
train2 = train_genre_df.copy()
train2 = train2.drop(dropColumns, axis=1)
train2.head()
```

```python
train3 = pd.merge(train2, users, on='UserID', how='outer')
train3.head()
```

```python
drop_indices = []
for ind, row in train3.iterrows():
    if math.isnan(row['Rating']):
        drop_indices.append(ind)
print drop_indices

train3 = train3.drop(train3.index[drop_indices])
print train3.isnull().values.any()
```

```python
train4 = train3.copy()
del train4['UserID']
del train4['MovieID']
train4.head()
```

```python
cols = train4.columns
cols = cols.drop('Unnamed: 0')
feature_cols = cols.drop('Rating')
print feature_cols

X4 = train4[feature_cols]
Y4 = train4['Rating']


mlreg4 = LinearRegression()
X_train4, X_test4, Y_train4, Y_test4 = train_test_split(X4, Y4, test_size=0.33, random_state=1)
mlreg4.fit(X_train4, Y_train4)
Y_pred4 = mlreg4.predict(X_test4)
```

```python
print np.sqrt(metrics.mean_squared_error(Y_test4, Y_pred4))
print metrics.mean_squared_error(Y_test4, Y_pred4)
```
