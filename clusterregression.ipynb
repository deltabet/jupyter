```python
import numpy as np
import pandas as pd

data = pd.read_csv('train_cat.csv')
data.head()
```

```python
data2 = data.copy()[['UserID', 'MovieID', 'Rating', 'Group', 'MovieType']]
data2.head()
```

```python
users = pd.read_csv('users_catergory.csv')[['UserID', 'Type2', 'Average']]
users.head()
```

```python
data3 = pd.merge(data2, users, on='UserID', how='outer')
data3['UserType'] = data3['Type2']
del data3['Type2']
data3['UserAvg'] = data3['Average']
del data3['Average']

data3.head()
```

```python
#train with just this
#add movieavg
#add genres/occupation
#add both
trainOld = pd.read_csv('trainGenre2.csv')
trainOld.head()
```

```python
userTypes = users.copy()
userTypes['UserTypes'] = userTypes['Type2']
del userTypes['Average']
del userTypes['Type2']
userTypes.head()
```

```python
train0 = pd.merge(trainOld, userTypes, on='UserID', how='outer')
del train0['Unnamed: 0']
del train0['Unnamed: 0.1']
print train0.shape
train0.head()
```

```python
data4 = data3.copy()[['UserID', 'MovieID', 'Group']]
print data4.shape
data4.head()
```

```python
train1 = pd.merge(train0, data4, on = ['UserID', 'MovieID'], how='left')
print train1.shape
train1.head()
```

```python
movies = pd.read_csv('movies_catergory.csv')[['MovieID', 'Type2']]
movies['MovieType'] = movies['Type2']
del movies['Type2']
movies.head()
```

```python
train2 = pd.merge(train1, movies, on='MovieID', how='outer')
train2.head()
```

```python
print train2.columns.values
```

```python
# figure out where the NaN values are and drop their rows
import math
print train2.shape

drop_indices = []
for ind, row in train2.iterrows():
    if math.isnan(row['Rating']):
        drop_indices.append(ind)
print drop_indices

train2 = train2.drop(train2.index[drop_indices])
print train2.isnull().values.any()
```

```python
trainCluster = train2.copy()[['UserID', 'MovieID', 'Rating', 
                            'Intercept', 'MovieAvg', 'UserAvg', 'UserTypes', 'Group', 'MovieType']]

del trainCluster['UserID']
del trainCluster['MovieID']
trainCluster.head()
```

```python
trainCluster2 = trainCluster.copy()
trainCluster2 = pd.concat([trainCluster2, pd.get_dummies(trainCluster2['MovieType'], prefix='M')], axis=1)
trainCluster2 = pd.concat([trainCluster2, pd.get_dummies(trainCluster2['UserTypes'], prefix='U')], axis=1)
del trainCluster2['MovieType']
del trainCluster2['UserTypes']
trainCluster2.head()
```

```python
from sklearn.cross_validation import train_test_split
from sklearn import metrics
from sklearn.linear_model import LinearRegression

cols = trainCluster2.columns
feature_cols = cols.drop('Rating')
print feature_cols

X = trainCluster2[feature_cols]
Y = trainCluster2['Rating']


clusterReg2 = LinearRegression()
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=1)
clusterReg2.fit(X_train, Y_train)
Y_pred = clusterReg2.predict(X_test)

print np.sqrt(metrics.mean_squared_error(Y_test, Y_pred))
print metrics.mean_squared_error(Y_test, Y_pred)
```

```python
zip(feature_cols, clusterReg2.coef_)
```
