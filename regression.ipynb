```python
movieInfo = movies.copy()
del movieInfo['Unnamed: 0']
del movieInfo['Title']
genres = ['Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime',
          'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical',
          'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
for genre in genres:
    movieInfo[genre] = np.zeros(len(movieInfo))
    
movieInfo['numGenres'] = np.zeros(len(movieInfo))
for i in range(len(movieInfo)):
    genreString = movieInfo.ix[i]['Genres']
    genreArr = genreString.split('|')
    count = 0
    for j in genreArr:
        if j in movieInfo:
            movieInfo.ix[i, j] = 1
            count = count + 1
        else:
            print i
            print j
            count = count + 1
    movieInfo.ix[i,'numGenres'] = count
movieInfo.head()
```

```python
import time
start = time.clock()
for i in range(1, len(userMovies)):
    ratings = ratingsArr.ix[:,i].as_matrix()
    for genre in genres:
        #print ratingsArr.ix[:,genre].as_matrix()
        product = ratings * ratingsArr.ix[:,genre].as_matrix()
        avgProduct = product[product != 0]
        if (len(avgProduct > 0)):
            avg = np.mean(avgProduct)
        else:
            avg = 0
        userMovies.ix[i - 1, genre] = avg
    if (i % 100 == 0):
        print i
end = time.clock() 
print (end - start)
userMovies.head()
```
